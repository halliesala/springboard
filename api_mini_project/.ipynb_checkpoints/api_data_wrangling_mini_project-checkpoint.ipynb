These are your tasks for this mini project:

1. Collect data from the Franfurt Stock Exchange, for the ticker AFX_X, for the whole year 2017 (keep in mind that the date format is YYYY-MM-DD).
2. Convert the returned JSON object into a Python dictionary.
3. Calculate what the highest and lowest opening prices were for the stock in this period.
4. What was the largest change in any one day (based on High and Low price)?
5. What was the largest change between any two days (based on Closing Price)?
6. What was the average daily trading volume during this year?
7. (Optional) What was the median trading volume during this year. (Note: you may need to implement your own function for calculating the median.)

```python
url = "https://data.nasdaq.com/api/v3/datasets/FSE/AFX_X.json"
params = dict(key=API_KEY, start_date='2017-01-01', end_date='2017-12-31')
res = requests.get(url, params)
```

```python
dataset = dict(res.json())['dataset']
```

```python
dataset.keys()
```

```python
cols = dataset['column_names']
cols
```

```python
data = dataset['data']
data[0]
```

```python
len(data)
```

```python
# create a list of dictionaries with cols:data[i] key-value pairs
afx_17_list = [dict(zip(cols, data[i])) for i in range(len(data))]

# create dict of dicts with dates as keys and dictionaries with cols:data[i] key-value pairs as values
afx_17_dict = {data[i][0]:dict(zip(cols, data[i])) for i in range(len(data))}
```

```python
# returns the median of an unsorted list
def median(vals):
    vals.sort()
    if len(vals) % 2 == 1:
        return vals[len(vals) // 2]
    else:
        return (vals[len(vals) // 2] + vals[(len(vals) // 2) - 1]) / 2
```

```python
# 3. Calculate what the highest and lowest opening prices were for the stock in this period.
# 4. What was the largest change in any one day (based on High and Low price)?
# 5. What was the largest change between any two days (based on Closing Price)?
# 6. What was the average daily trading volume during this year?


highest = float('-inf')
lowest = float('inf')
largest_change = 0
largest_change_2day = 0
yesterday = None
trade_vol = []


for date in afx_17_dict.keys():
    # update the highest, lowest open; skip 'None' values
    try:
        highest = max(highest, afx_17_dict[date]['Open'])
        lowest = min(lowest, afx_17_dict[date]['Open'])
    except TypeError:
        pass
    
    # update the largest 1-day change
    largest_change = max(largest_change, abs(afx_17_dict[date]['High']-afx_17_dict[date]['Low']))
    
    # update the largest 2-day change
    try: 
        largest_change_2day = max(largest_change_2day, abs(afx_17_dict[date]['Close'] - afx_17_dict[yesterday]['Close']))
    except KeyError:
        pass
    yesterday = date
    
    # update the list of daily trading volume
    trade_vol.append(afx_17_dict[date]['Traded Volume'])
    
    
print("Highest open:", highest)
print("Lowest open:", lowest)
print("Largest one-day change (high-low):", largest_change)
print("Largest two-day change (close):", largest_change_2day)
print("Average daily trading volume:", sum(trade_vol)/len(trade_vol))
print("Median daily trading volume:", median(trade_vol))
```
